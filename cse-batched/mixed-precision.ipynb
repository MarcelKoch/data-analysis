```python
def read(data: str):
    return pd.read_csv(io.StringIO(data), sep="|", header=None, names = ["d1", "N", "dt", "err", "order", "d2"], dtype={"N":int, "dt": float, "err": float, "order": float}).drop(columns=["d1", "d2"]).set_index("dt")
```
